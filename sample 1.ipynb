```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
data = pd.read_csv('sample.csv')
```

```python
plt.scatter(data['SAT'], data['GPA'])
plt.xlabel('SAT')
plt.ylabel('GPA')
```

```python
x = preprocessing.scale(data)
kmeans = KMeans(4)
kmeans.fit(x)
```

```python
plt.scatter(data['SAT'], data['GPA'],c=kmeans.fit_predict(x),cmap='rainbow')
plt.xlabel('SAT')
plt.ylabel('GPA')
```
